```python
#question 2
def is_even(num):
  if num%2 ==0:
    return True
  else:
    return False

if is_even(float(input("Enter a number here: "))):
  print("The given number is Even")
else:
  print("The given number is Odd")
```


**Even Number Check**:
\
The code defines a function named *is_even *that takes a single argument *num*.
*   Inside the is_even function, it uses an if statement to check if the remainder of num divided by 2 is equal to 0.
*   If the remainder is 0, it means the number is divisible by 2, so it's even.\
The function returns True in this case.
*   If the remainder is not 0, the number is odd.
\
The function returns False in this case.








```python
#question 3
def convert_temperature(temp, scale):
  if scale == 'celcius' or scale.upper() == 'C':
    return (temp * 9/5) + 32 , 'F'
  elif scale == 'fahrenheit' or scale.upper() == 'F':
    return (temp - 32) * 5/9 , 'C'
  else:
    print("Invalid input ")
    return None , None

temp= float(input("Enter the temperature: "))
scale = input("Enter the scale (celcius or fahrenheit): ")
con_temp, con_scale=convert_temperature(temp, scale)
print(str(temp)+" "+scale.upper()+" = "+ str(con_temp)+" "+con_scale)
```

***Function convert_temperature()***

*   Takes two arguments: temp (the temperature value) and scale (the original temperature scale).


*   Checks the scale input (case-insensitive) using if and elif statements:
*   If it's "celsius" or uppercase "C", converts to Fahrenheit using the formula (temp * 9/5) + 32.
*   If it's "fahrenheit" or uppercase "F", converts to Celsius using the formula (temp - 32) * 5/9.
*   In both cases, it returns a tuple containing the converted temperature and the new scale ("F" or "C").
*   If the scale input doesn't match "celsius" or "fahrenheit" (ignoring case), it prints an "Invalid input" message and returns None for both converted temperature and scale.


```python
#question 4
def find_max_min(numbers_list):
  max,min= numbers_list[0], numbers_list[0]
  for i in range(len(numbers_list)):
    if max<=numbers_list[i]:
      max=numbers_list[i]
    if min>=numbers_list[i]:
      min=numbers_list[i]
  return max,min

num_list=[]
print("Enter 5 numbers of your choice: ")
for i in range(5) :
  num_list.append(int(input()))
max,min=find_max_min(num_list)
print(f"Maximum number is: {max} ")
print(f"Minimum number is: {min}")
```

***Function find_max_min:***
\
**Initializes max and min:**\
 Sets both max and min to the first element of the numbers_list (assuming all elements are numbers).
\
**Iterates through the list:**\
 Uses a for loop to iterate through each element (numbers_list[i]) in the list.
\
**Compares elements:**

*   Checks if the current element (numbers_list[i]) is greater than or equal
to the current max.
*   If yes, updates max to the current element.

*  Checks if the current element is less than or equal to the current min.

*  If yes, updates min to the current element.



**Returns max and min:** \
After iterating through the entire list, the function returns both the final max and min values.

```python
#question 6
def update_inventory(inventory_dict, item, quantity,opt):
  if item in inventory_dict:
    if opt.lower()=="add":
      inventory_dict[item]+=quantity
    elif opt.lower()=="remove" and (inventory_dict[item]-quantity)>0:
      inventory_dict[item]-=quantity
    else:
      inventory_dict[item]=quantity

  else:
    inventory_dict[item]=quantity


dic={"icecream":2, "lipton":5, "colgate":4,"burger":8,"bear":0}
print(dic)

print("Update this dictionary")

item1=input("Item: ")
quantity1=int(input("Quantity: "))
opt1=input("Option(Add/Remove): ")
update_inventory(dic,item1,quantity1,opt1)

item2=input("Item: ")
quantity2=int(input("Quantity: "))
opt2=input("Option(Add/Remove): ")
update_inventory(dic,item2,quantity2,opt2)

item3=input("Item: ")
quantity3=int(input("Quantity: "))
opt3=input("Option(Add/Remove): ")
update_inventory(dic,item3,quantity3,opt3)

print("The updated dictionary is: ")
print(dic)
```

\**Function Input:**
\
***inventory_dict***: The dictionary holding current stock levels (item names as keys, quantities as values).\
***item***: The item name to be added or removed.\
***quantity***: The quantity to be added/removed (positive for adding, negative for removing).\
***opt***: The operation to perform ("Add" or "Remove").
\
**Checking Existing Item:**
\
 **If the item already exists in the inventory_dict:**\

*   For "Add" operation (opt.lower() == "add"): Adds the quantity to the existing stock level.
*   For "Remove" operation (opt.lower() == "remove") with sufficient stock:\
Checks if the desired removal amount (quantity) doesn't exceed existing stock (inventory_dict[item] - quantity > 0).
*   If there's enough stock, subtracts the quantity from the existing amount.\
Otherwise (invalid "Remove" or insufficient stock): Sets the stock level for that item to the desired quantity.

**Adding new item**:
\
If the item is not found in the dictionary, it's treated as a new item and added to the dictionary with the provided quantity.
